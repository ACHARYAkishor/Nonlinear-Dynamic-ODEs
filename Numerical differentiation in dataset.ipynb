# Compute Fintite difference derivative

Lets compute the finiite difference derivative form the data, whihc is a synthetatic data generated by $Sin(x)$

```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use(['science', 'notebook', 'grid'])
```

#### Synthetatic Dataset

```python
# Creating synthetatic dataset

n = 30 # number of datapoints (make it corse or fine by changing the value)
x = np.linspace(0.1,3,n) # a position vector
f = np.sin(x) # dataset vector

#plotting the data
plt.plot(x,f,'k') #plotting true function
plt.plot(x,f,'rx', linewidth=2) # plotting the datapoints
plt.show()
```

### Finite Difference Derivative

```python

dx = x[1] - x[0] # the difference
dfdx = np.zeros(n) # definitng an empty vector
```

```python
# perfomring the finite difference 

dfdx[0] = (f[1] - f[0])/(x[1] - x[0]) # forward difference at f(x_1)

for i in range(1,n-1):
    dfdx[i] = (f[i+1] - f[i-1])/(x[i+1]-x[i-1]) #central difference

dfdx[-1] = (f[-1] - f[-2])/(x[-1]-x[-2]) # backward difference
```
