# Numerical Computation Schemes Comparisions

How to approximate $\textbf{First derivatives (df/dt)}$  by using finite difference derivatives?
We are goinig to compare three numerical schemes:

* Forward difference
* Backward difference
* Central difference

```python
import numpy as np
from matplotlib import pyplot as plt

plt.style.use(['science', 'notebook', 'grid'])
```

### Exact Derivative

```python
t = np.linspace(-4,4,80) # for plotting the known function
f = np.sin(t) # a known function

#Exact derivative of known function
dfdt = np.cos(t)
```

```python
#plot

plt.plot(t,f, 'k--', linewidth=1.2) # plotting known function
plt.plot(t, dfdt, 'k', linewidth=3) #plotting the known derivative
plt.legend(['Function', 'Exact Derivative'])
plt.xlim(-4,4)
plt.ylim(-1.5,1.5)
plt.show()
```

### Approximate derivatives

First we will define a small time step $dt$ and then try to study how the error behaves while changing the $dt$ for different schemes.

```python
dt = 0.2 # define for numerical computation
# change the dt (bigger and smaller) and see how approximation error changes for each schemes
```

#### Forward difference scheme $O(dt)$

$$
\frac{df}{dt} = \frac {f(t+dt) - f(t)}{dt}
$$

```python
# Forward difference Approximation
dfdt_F = (np.sin(t + dt) - np.sin(t))/dt
```

#### Backward difference scheme $O(dt)$

$$
\frac{df}{dt} = \frac {f(t)- f(t+dt)}{dt}
$$

```python
# Backward difference Approximation
dfdt_B = (np.sin(t) - np.sin(t-dt))/dt
```

#### Central difference scheme $O(dt^{2})$

$$
\frac{df}{dt} = \frac {f(t + dt)- f(t-dt)}{2dt}
$$

```python
# Central difference Approximation
dfdt_C = (np.sin(t + dt) - np.sin(t-dt))/ (2* dt)
```

```python
# plot

plt.figure(figsize=(20 ,4))
plt.subplot(1,2,1)
plt.plot(t, dfdt, 'k', linewidth=3) # exact derivative
plt.plot(t, dfdt_F, 'b', linewidth=1.2) # forward difference
plt.plot(t, dfdt_B, 'g', linewidth=1.2) # backward difference
plt.plot(t, dfdt_C, 'r', linewidth=1.2) # central difference
plt.legend(['Exact difference', 'Forward difference', 'Backward difference', 'Central difference'], fontsize=14, framealpha=1.0)

# zoom in plot
plt.subplot(1,2,2)
plt.plot(t, dfdt, 'k', linewidth=3) # exact derivative
plt.plot(t, dfdt_F, 'b', linewidth=1.2) # forward difference
plt.plot(t, dfdt_B, 'g', linewidth=1.2) # backward difference
plt.plot(t, dfdt_C, 'r', linewidth=1.2) # central difference
plt.xlim(0,2)
plt.legend(['Exact difference', 'Forward difference', 'Backward difference', 'Central difference'], fontsize=14, framealpha=1.0)
```
